```python
df3['TransactionDate'] = pd.to_datetime(df3['TransactionDate'], errors='coerce')
df3['Hour'] = df3['TransactionDate'].dt.hour

def time_slot(hour):
  if 0 < hour < 3:
    return '12am - 3am'
  elif 3 <= hour < 6:
    return '3am - 6am'
  elif 6 <= hour < 9:
    return '6am - 9am'
  elif 9 <= hour < 12:
    return '9am - 12pm'
  elif 12 <= hour < 15:
    return '12pm - 3pm'
  elif 15 <= hour < 18:
    return '3pm - 6pm'
  elif 18 <= hour < 21:
    return '6pm - 9pm'
  elif 21 <= hour < 24:
    return '9pm - 12am'
```

```python
df3['Time Slot'] = df3['Hour'].apply(time_slot)
```

```python
active_time_period = df3.groupby('Time Slot')['CustomerID'].nunique()
active_time_period = active_time_period.sort_values(ascending=False)
```

```python
plt.figure(figsize=(4,4))
active_time_period.plot(kind='bar')

plt.title('CUSTOMER ACTIVITY BY TIME OF THE DAY')
plt.ylabel('No.Of Customers')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
customers =pd.read_csv('Customers.csv')
product=pd.read_csv('Products.csv')
transaction=pd.read_csv('Transactions.csv')
```

```python
merged_df = pd.merge(transaction, customers, on='CustomerID', how='inner')
```

```python
# Exploratory Data Analysis (EDA)
print("Summary of Transactions:")
print(merged_df.describe())
```

```python
merged_df.columns
```

```python
# Total revenue per customer
customer_revenue = merged_df.groupby('CustomerID')['Price'].sum().reset_index()
customer_revenue.columns = ['CustomerID', 'TotalSpent']
```

```python
customer_revenue
```

```python
# Frequency of purchases per customer
customer_frequency = merged_df.groupby('CustomerID').size().reset_index(name='Frequency')
customer_frequency
```

```python
# Recency: Days since last purchase
merged_df['TransactionDate'] = pd.to_datetime(merged_df['TransactionDate'])
last_purchase = merged_df.groupby('CustomerID')['TransactionDate'].max().reset_index()
last_purchase['Recency'] = (merged_df['TransactionDate'].max() - last_purchase['TransactionDate']).dt.days
```

```python
last_purchase
```

```python
# Merge all customer metrics
customer_data = customer_revenue.merge(customer_frequency, on='CustomerID').merge(last_purchase[['CustomerID', 'Recency']], on='CustomerID')
customer_data
```

```python
# Standardize the data for clustering
scaler = StandardScaler()
customer_data_scaled = scaler.fit_transform(customer_data[['TotalSpent', 'Frequency', 'Recency']])
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(customer_data_scaled)
```

```python
customer_data
```

```python
# Visualizing customer segments
plt.figure(figsize=(10, 6))
sns.scatterplot(x=customer_data['TotalSpent'], y=customer_data['Frequency'], hue=customer_data['Cluster'], palette='viridis')
plt.xlabel('Total Amount Spent')
plt.ylabel('Purchase Frequency')
plt.title('Customer Segmentation')
plt.show()
```

```python
# Save segmented data
customer_data.to_csv("Customer_Segments.csv", index=False)
print("Customer segmentation completed and saved to Customer_Segments.csv")
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Load datasets
customers = pd.read_csv("customers.csv")
products = pd.read_csv("products.csv")
transactions = pd.read_csv("transactions.csv")
```

```python
# Extract first 20 customer IDs
target_customers = customers.iloc[:20]['CustomerID'].tolist()
target_customers
```

```python
# Merge transactions with products
txn_merged = transactions.merge(products, on="ProductID")

# Create a customer-product matrix
customer_product_matrix = txn_merged.pivot_table(index='CustomerID', 
                                                 columns='ProductID', 
                                                 values='Quantity', 
                                                 aggfunc='sum', 
                                                 fill_value=0)

customer_product_matrix
```

```python
# Standardize the matrix
scaler = StandardScaler()
customer_product_matrix_scaled = scaler.fit_transform(customer_product_matrix)
```

```python
# Compute cosine similarity
similarity_matrix = cosine_similarity(customer_product_matrix_scaled)
similarity_df = pd.DataFrame(similarity_matrix, 
                             index=customer_product_matrix.index, 
                             columns=customer_product_matrix.index)

similarity_df
```

```python
# Find top 3 similar customers for each target customer
lookalike_map = {}
for cust_id in target_customers:
    similar_customers = similarity_df.loc[cust_id].drop(cust_id).nlargest(3)
    lookalike_map[cust_id] = list(zip(similar_customers.index, similar_customers.values))

# Convert to DataFrame
lookalike_df = pd.DataFrame([(k, v[0], v[1]) for k, vals in lookalike_map.items() 
                             for v in vals],
                             columns=["CustomerID", "LookalikeID", "Score"])

lookalike_df
```

```python
# Save to CSV
lookalike_df.to_csv("Lookalike.csv", index=False)
print("Lookalike.csv has been created successfully!")
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Load datasets
customers = pd.read_csv("customers.csv")
products = pd.read_csv("products.csv")
transactions = pd.read_csv("transactions.csv")

# Extract first 20 customer IDs
target_customers = customers.iloc[:20]['CustomerID'].tolist()

# Merge transactions with products
txn_merged = transactions.merge(products, on="ProductID")

# Create a customer-product matrix
customer_product_matrix = txn_merged.pivot_table(index='CustomerID', 
                                                 columns='ProductID', 
                                                 values='Quantity', 
                                                 aggfunc='sum', 
                                                 fill_value=0)

# Standardize the matrix
scaler = StandardScaler()
customer_product_matrix_scaled = scaler.fit_transform(customer_product_matrix)

# Compute cosine similarity
similarity_matrix = cosine_similarity(customer_product_matrix_scaled)
similarity_df = pd.DataFrame(similarity_matrix, 
                             index=customer_product_matrix.index, 
                             columns=customer_product_matrix.index)

# Find top 3 similar customers for each target customer
lookalike_map = {}
for cust_id in target_customers:
    similar_customers = similarity_df.loc[cust_id].drop(cust_id).nlargest(3)
    lookalike_map[cust_id] = list(zip(similar_customers.index, similar_customers.values))

# Convert to DataFrame
lookalike_df = pd.DataFrame([(k, v[0], v[1]) for k, vals in lookalike_map.items() for v in vals],
                             columns=["CustomerID", "LookalikeID", "Score"])

# Save to CSV
lookalike_df.to_csv("Lookalike2.csv", index=False)
print("Lookalike.csv has been created successfully!")

# Visualize the data using a graph
G = nx.Graph()
for cust_id, lookalikes in lookalike_map.items():
    for lookalike_id, score in lookalikes:
        G.add_edge(cust_id, lookalike_id, weight=score)

plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', edge_color='gray')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Customer Lookalike Network")
plt.show()
```
